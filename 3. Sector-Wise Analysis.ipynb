```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Spending data.xls",skiprows=[0],header=1)
```

### Sector-Wise Analysis:

##### To conduct a sector-wise analysis and compare government spending trends across different sectors, I used folowing steps:

**Data Preparation:**
- Grouped the data by sectors.
- Aggregated spending or commitment amounts for each sector over time.

**Visualization:**
- Ploted bar plots to compare spending trends across sectors.
- Identify any significant fluctuations or anomalies in specific sectors.

**Statistical Analysis:**
- Calculate descriptive statistics for each sector to compare mean, median, and standard deviation.
- Conduct trend analysis or growth rate comparisons across sectors.
uct trend analysis or growth rate comparisons across sectors.

#### Sector 1:-

```python
sector1_spending = df.groupby('sector1')['curr_total_commitment'].sum()

# Sorting sectors by total commitment amount in descending order
sector1_spending = sector1_spending.sort_values(ascending=False)

# Display top sectors with highest total commitment amounts
sector1_spending.head(10)
```

```python
plt.figure(figsize=(10, 6))
sector1_spending.head(10).plot(kind='bar', color='skyblue')
plt.title('Top Sectors by Total Commitment Amount')
plt.xlabel('Sectors')
plt.ylabel('Total Commitment Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sector1_statistics = df.groupby('sector1')['curr_total_commitment'].describe()

# Display statistics for top sectors
sector1_statistics.loc[sector1_spending.head(10).index]
```
